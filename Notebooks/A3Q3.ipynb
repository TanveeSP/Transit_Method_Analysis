```python
# Constants for Jupiter
R_jupiter = 69911 * u.km  # Radius of Jupiter
distance_jupiter_sun = 5.2 * u.au  # Distance of Jupiter from the Sun
distance_probe_sun = 50 * u.au  # Distance of the space probe from the Sun

#  distance from probe to Jupiter (in AU and convert to km)
distance_probe_jupiter = (distance_probe_sun - distance_jupiter_sun).to(u.km)

# Calculating the angular size in radians
theta_jupiter_rad = 2 * np.arctan(R_jupiter / distance_probe_jupiter)

# angular size to degrees
theta_jupiter_deg = theta_jupiter_rad.to(u.deg)

print(f"Angular size of Jupiter from the space probe: {theta_jupiter_rad:.4e} radians")
print(f"Angular size of Jupiter from the space probe: {theta_jupiter_deg:.4f} degrees")
```
