```python
import numpy as np
import astropy.units as u

# Constants for the Sun and Jupiter
R_sun = 696340 * u.km  # Radius of Sun
R_jupiter = 69911 * u.km  # Radius of Jupiter
R_earth = 6371 * u.km  # Radius of Earth
F_sun = 1370 * u.W / u.m**2  # Flux of Sun at Earth

# Distance of the space probe from the Sun (1 parsec)
d_probe_sun = 1 * u.pc  # 1 parsec = 3.086 x 10^13 km

# a. Flux received by the space probe
flux_probe = F_sun * (1 * u.AU / d_probe_sun)**2
print(f"Flux received by the space probe: {flux_probe:.2e}")

# b. Angular size of the Sun from the space probe
theta_sun_rad = 2 * np.arctan(R_sun / d_probe_sun)
theta_sun_deg = theta_sun_rad.to(u.deg)
print(f"Angular size of the Sun from the space probe: {theta_sun_deg:.4f}")

# c. Angular size of Jupiter from the space probe (distance , probe to Jupiter ≈ 1 pc)
theta_jupiter_rad = 2 * np.arctan(R_jupiter / d_probe_sun)
theta_jupiter_deg = theta_jupiter_rad.to(u.deg)
print(f"Angular size of Jupiter from the space probe: {theta_jupiter_deg:.4f}")

# d.  Fraction of the Sun covered by Jupiter
fraction_covered_jupiter = (R_jupiter / R_sun)**2
print(f"Fraction of the Sun covered by Jupiter: {fraction_covered_jupiter:.4e}")

# e. Change in magnitude due to Jupiter's transit
F_final_jupiter = F_sun * (1 - fraction_covered_jupiter)
delta_m_jupiter = -2.5 * np.log10(F_final_jupiter / F_sun)
print(f"Change in magnitude due to Jupiter's transit: {delta_m_jupiter:.4f}")

# f. Change in magnitude due to Earth's transit
fraction_covered_earth = (R_earth / R_sun)**2
F_final_earth = F_sun * (1 - fraction_covered_earth)
delta_m_earth = -2.5 * np.log10(F_final_earth / F_sun)
print(f"Change in magnitude due to Earth's transit: {delta_m_earth:.4f}")
```
