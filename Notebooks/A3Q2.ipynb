```python
import numpy as np
from astropy import units as u

# Constants
R_sun = 696340 * u.km  # Radius of the Sun
distance_probe = 50 * u.au  # Distance from the Sun to the space probe

#  convert AU to kilometers
distance_probe_km = distance_probe.to(u.km)

# Calculating angular size in radians
theta_rad = 2 * np.arctan(R_sun / distance_probe_km)

#  angular size to degrees
theta_deg = theta_rad.to(u.deg)

print(f"Angular size of the Sun from the space probe: {theta_rad:.4e} radians")
print(f"Angular size of the Sun from the space probe: {theta_deg:.4f} degrees")
```
