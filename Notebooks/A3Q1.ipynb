```python
import numpy as np
from astropy import units as u
from astropy.constants import au

# Values Given
F_Earth = 1370 * u.W / u.m**2  # Flux at Earth's distance
R_Earth = 1 * au  # Distance from the Sun to Earth
R_probe = 50 * au  # Distance from the Sun to the space probe
F_0 = 2.51802e-8 * u.W / u.m**2  # Flux corresponding to zero magnitude

# 1: flux received by the space probe
F_probe = F_Earth * (R_Earth / R_probe)**2

# 2: bolometric magnitude of the Sun
m_bol = -2.5 * np.log10(F_probe / F_0)

print(f"Flux received by the space probe: {F_probe:.4e}")
print(f"Bolometric magnitude of the Sun: {m_bol:.2f}")
```
